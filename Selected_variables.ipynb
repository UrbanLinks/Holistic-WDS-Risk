```python
import pandas as pd
import re
```

```python
Data = pd.read_csv('test-python.csv', encoding='latin-1',
                   dtype={'Cities': str}, low_memory=False)

pattern = r'^(a|b|c|d|e|f|g|h|i|j|k|l|m|n|o|p|q|r|s|t|u|v|w|x|y|z|ab|ac|ad|af|ag|ah|ai|aj|ak|al|am|an|ao)$'
filtered_data = Data[Data['CHARACTERISTIC_NAME'].astype(str).str.contains(pattern, flags=re.IGNORECASE, regex=True, na=False)]
display(filtered_data)
```
